# Inflacion desde World Bank/WDI

Este notebook descarga inflacion anual medida por precios al consumidor.

## 1. Importar la API de datos

```python
from economista import data
```

## 2. Descargar inflacion anual

El indicador `FP.CPI.TOTL.ZG` mide inflacion anual de precios al consumidor.

```python
data_inflation = data.fetch(
    source="world_bank",
    dataset="wdi",
    indicator="FP.CPI.TOTL.ZG",
    geo=["MEX", "COL", "ARG", "GBR"],
    start=2000,
    end=2024,
)

data_inflation.metadata.indicator_name
```

## 3. Comparar paises

El valor ya viene como porcentaje anual. Redondeamos a un decimal para evitar ruido visual.

```python
df_inflation = data_inflation.to_pandas()

inflation_percent = (
    df_inflation.pivot(index="time", columns="geo", values="value")
    .sort_index()
)

inflation_percent.tail().round(1)
```
