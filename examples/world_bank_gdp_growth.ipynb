# Crecimiento del PIB desde World Bank/WDI

Este notebook descarga crecimiento anual del PIB real y compara paises en puntos porcentuales.

## 1. Importar la API de datos

```python
from economista import data
```

## 2. Descargar crecimiento anual del PIB

El indicador `NY.GDP.MKTP.KD.ZG` mide el crecimiento anual del PIB real.

```python
data_growth = data.fetch(
    source="world_bank",
    dataset="wdi",
    indicator="NY.GDP.MKTP.KD.ZG",
    geo=["MEX", "COL", "ARG", "GBR"],
    start=2000,
    end=2024,
)

data_growth.metadata.indicator_name
```

## 3. Comparar paises

El valor ya viene como porcentaje anual. Solo redondeamos para lectura.

```python
df_growth = data_growth.to_pandas()

growth_percent = (
    df_growth.pivot(index="time", columns="geo", values="value")
    .sort_index()
)

growth_percent.tail().round(1)
```
