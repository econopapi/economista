# Poblacion desde World Bank/WDI

Este notebook descarga poblacion total y la muestra en millones de personas.

## 1. Importar la API de datos

```python
from economista import data
```

## 2. Descargar poblacion total

El indicador `SP.POP.TOTL` representa poblacion total.

```python
data_population = data.fetch(
    source="world_bank",
    dataset="wdi",
    indicator="SP.POP.TOTL",
    geo=["MEX", "COL", "USA", "BRA"],
    start=2000,
    end=2024,
)

data_population.metadata.indicator_name
```

## 3. Convertir a millones

World Bank entrega personas. Para lectura rapida, usamos millones de personas.

```python
df_population = data_population.to_pandas()

population_millions = (
    df_population.assign(value_millions=df_population["value"] / 1_000_000)
    .pivot(index="time", columns="geo", values="value_millions")
    .sort_index()
)

population_millions.tail().round(1)
```
